```python
import project_functions
import pandas as pd
import seaborn as sns
```

```python
df = project_functions.load_and_process('../../data/raw/project_data.csv')
```

#### Heatmap correlation and mean attributes grouped by win/loss ####

```python
    sns.heatmap(df.corr() , xticklabels=df.corr().columns, yticklabels=df.corr().columns, annot=True, cmap=sns.diverging_palette(220, 10, as_cmap=True)).set_title("Correlation of columns in dataset")
```

```python
df.groupby('won', as_index = False).mean()
```

### **First EDA: Relationship between shots and the result (win/loss)** ###

```python
sns.barplot(y = 'shots', x= 'won', data =df).set_title("shots by winning/losing teams")
```

```python
sns.stripplot(y = 'shots', x ='won', data = df).set_title("shots by winning/losing teams")
```

```python
sns.kdeplot(x = 'shots', hue = 'won', data = df).set_title("shots by winning/losing teams")
```

**No significant difference between shot total and the result of the game**

### **Second EDA: Do home teams generally have more puck possesion? (Faceoff percentage)** ###

```python
sns.displot(x = 'faceoffs(%)', hue = 'home/away', data = df)
```

```python
sns.stripplot(x = 'faceoffs(%)', y = 'home/away', data = df).set_title("faceoff percentage by home/away teams")
```

```python
    sns.barplot(y = 'faceoffs(%)', x = 'home/away', data = df).set_title("faceoff percentage by home/away teams")
```

**Home teams have better faceoff percentages (control more of the puck)**

### **Third EDA: Do winning teams generally have a good powerplay? (And vice versa)** ###

```python
sns.kdeplot(x = 'ppg', hue = 'won', data = df).set_title("powerplay goals for winning/losing teams")
```

```python
sns.violinplot(y = 'ppg', x = 'won', data = df).set_title("powerplay goals for winning/losing teams")
```

```python
sns.barplot(y = 'ppg', x = 'won', data = df).set_title("powerplay goals for winning/losing teams")
```

**Winning teams generally score more powerplay goals than losing teams. Notice the large spike at 0 goals for the losing team on the kdeplot**
